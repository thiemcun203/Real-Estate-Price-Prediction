```python
with open('total_dict.pickle', 'rb') as f:
    loc_data = pickle.load(f)
print(loc_data)
```

```python
def get_loc(address, loc_dict=loc_data):
    return loc_dict[address] if address in loc_dict and loc_dict[address]!=(float('nan'), float('nan')) else (float('nan'), float('nan'))
```

```python
data['Address']=data['road']+','+data['ward']+','+data['district']+" hà nội"
data.head()
```

```python
final_data = data.copy()
final_data['X'], final_data['Y'] = zip(*final_data['Address'].apply(get_loc))
```
