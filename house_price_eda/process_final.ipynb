## PROCESS DATA

### Processing legal

```python
house_data.loc[house_data['legal'] == 'Sổ hồng/ Sổ đỏ', 'legal'] = 'Sổ đỏ/ Sổ hồng'
house_data['legal'].unique()
```

### Processing housetype

```python
house_data['house_type'].unique()
```

```python
type_dict = {'Bán nhà ngõ, hẻm': 'Nhà ngõ, hẻm',
             'Nhà trong hẻm':'Nhà ngõ, hẻm',
             'Bán nhà mặt tiền': 'Nhà mặt tiền',
             'Bán căn hộ chung cư': 'Căn hộ chung cư',
                'Bán nhà riêng': 'Nhà riêng',
                'Bán biệt thự, villa,': 'Biệt thự, villa, nhà liền kề',
                'Biệt thự, nhà liền kề': 'Biệt thự, villa, nhà liền kề',
                'Biệt thự': 'Biệt thự, nhà liền kề',
                'Bán đất thổ cư': 'Đất thổ cư, đất ở',
                'Bán đất nền dự án': 'Đất nền, liền kề, đất dự án',
                'Bán nhà trọ':'Phòng trọ, nhà trọ',
                'Bán kho xưởng': 'Kho, xưởng',
                'Bán đất nông nghiệp':'Đất nông, lâm nghiệp',
                'Bán condotel':'Nhà hàng, khách sạn',
                'Bán Shophouse': 'Shop, kiot, quán', #vawn phong , mat bang, dat, datnen phanlo, 'Bán nhà cấp 4'
                'Bán loại bất động sản khác': 'Các loại khác',
                'Đất nền, phân lô': 'Đất nền, liền kề, đất dự án',
                'Đất':'Đất nền, liền kề, đất dự án',
                'Bán nhà cấp 4':'Nhà cấp 4',

}
```

```python
def replace_house_type(x, type_dict=type_dict):
    if x in type_dict.keys():
        return type_dict[x]
    else:
        return x
```

```python
house_data['house_type'] = house_data['house_type'].apply(replace_house_type)
```

```python
house_data['house_type'].unique()
```

### Process 'car_parking', 'rooftop_terrace', 'kitchen', 'dining_room'

```python
columns = ['car_parking', 'rooftop_terrace', 'kitchen', 'dining_room']

# Set values to 'Yes' where the condition is True
house_data.loc[:, columns] = house_data.loc[:, columns].applymap(lambda x: 'Yes' if x == True else x)
```

```python
house_data[columns].nunique()
```

```python
for col in house_data.columns:
    if col not in ['price', 'area', 'road_width', 'facade', 'length', 'storey', 'bedroom', 'toilet', 'X', 'Y', 'description', 'address']:
        print(col, house_data[col].unique().tolist())
```
